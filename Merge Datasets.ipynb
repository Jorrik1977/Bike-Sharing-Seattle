# Merge two final Datasets for Predictive Modeling

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
%matplotlib inline
```

```python
trips = pd.read_csv('trips_data_clean.csv', error_bad_lines=False, parse_dates=['date'], infer_datetime_format=True)
weather = pd.read_csv('weather_data_final1.csv', error_bad_lines=False, parse_dates=['date'], infer_datetime_format=True)
```

# Merge First Final

```python
# Merge trips and weather dataset on date
final = pd.merge(trips, weather, left_on="date", right_on="date")
pd.set_option('display.max_columns', 50) 
final.head()
```

```python
final.to_csv('/Users/jorrikfulda/nf-ds/capstone/finale.csv', index=False)
```

# Produce Second Final

```python
# count how many trips have been taken on a given day
# as we only have the weather data for a whole day
df = final.groupby(['date']).count()
df
```

```python
# reset date from being the index to being a feature
df.reset_index(level=0, inplace=True)
df
```

```python
# extract only date and count of trips
col_list = ['date', 'trip_id']
df = df[col_list]
df
```

```python
# rename feature
df.rename(columns={"trip_id": "cnt"})
```

```python
# merge weather info to this dataframe
final2 = pd.merge(df, weather, left_on="date", right_on="date")
pd.set_option('display.max_columns', 50) 
final2.head()
```

```python
# add new features 
final2['month'] = pd.DatetimeIndex(final2['date']).month # add month as column
final2['day'] = pd.DatetimeIndex(final2['date']).day
final2['weekday'] = (final2['day'] < 5).astype(float) # add weekday or not
#final2['season'] = 
final2.head()
```

```python
# export final csv
final2.to_csv('/Users/jorrikfulda/nf-ds/capstone/finale2.csv', index=False)
```
